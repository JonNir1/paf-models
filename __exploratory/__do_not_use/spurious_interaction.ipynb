```python
import numpy as np
import statsmodels.formula.api as smf

_SEED = 42
rng = np.random.default_rng(_SEED)
```

### True Model
$Y = 1 + 2*X1 + 3*X2 + e$
where $e \sim N(0, 1)$ and $X1, X2 \sim Ber(0.5)$ independent.

```python
n = 5000
X1 = rng.binomial(1, 0.5, n)
X2 = rng.binomial(1, 0.5, n)
e = rng.normal(0, 1, n)
Y = 1 + 2*X1 + 3*X2 + e
```

### Spurious Interaction Model
We will check if we can get a significant interaction in the model $Y = 1 + b*X1*X2 + e$, despite the fact that there is no interaction in the true model.

```python
model = smf.ols("Y ~ X1:X2", data={"Y": Y, "X1": X1, "X2": X2})
results = model.fit()
print(results.summary())
```
